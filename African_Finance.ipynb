```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('finance1.csv',encoding='latin-1')
```

```python
df.head()
```

```python
df.info()
```

```python
df1=df.dropna()
```

```python
df1.info()
```

```python
df1.rename(columns={'Starting Date':'starting_date'},inplace=True)
```

```python
df1.starting_date=pd.to_datetime(df1.starting_date)
```

```python
df1.info()
```

```python
import datetime as dt
df1['starting_year']=df1['starting_date'].dt.year
```

```python
df1.head()
```

```python
dff=df1[df1['starting_year']==2019]
```

```python
dff
```

```python
dff.drop(["Title","Project Code","starting_date","Status","Source of Financing","Sector","Sovereign"],axis=1,inplace=True)
```

```python
dff
```

```python
dff['Commitment in U.A'] = dff['Commitment in U.A'].str.replace(',', '').astype(float)
```

```python
total_commitments = dff.groupby('Country')['Commitment in U.A'].sum().reset_index()
```

```python
total_commitments['Commitment in U.A'].sum()
```

```python
dff['Commitment in U.A'].sum()
```

```python
total_commitments.plot(x='Country',y='Commitment in U.A',kind='bar',figsize=(8,8))
```
